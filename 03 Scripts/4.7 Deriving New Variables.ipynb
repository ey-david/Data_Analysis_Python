## 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled â€œMost orders,â€ â€œAverage orders,â€ and â€œFewest orders.â€ Create a new column containing these labels called â€œbusiest_period_of_day.â€

```python
# first see where the frequency lies
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
df_ords_prods_merged = df_ords_prods_merged.drop(columns = ['_merge'])
```

```python
# Now to separate them and create values
busiest_hours_result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value >= 9 and value <= 16:
    busiest_hours_result.append("Most orders")
  elif value == 7 or value == 8 or (value >= 17 and value <= 22):
    busiest_hours_result.append("Average orders")
  else:
    busiest_hours_result.append("Fewest orders")
```

```python
busiest_hours_result
```

```python
# Add the new column
df_ords_prods_merged['busiest_period_of_day'] = busiest_hours_result
```
